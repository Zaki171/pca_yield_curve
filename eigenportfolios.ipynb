```python
import numpy as np
import matplotlib.pyplot as plt



def calc_weights(df, lam=0.985):
    w_vec = []
    for i in range(len(df)):
        w_vec.append((1-lam)*(lam**i))
    w_vec.reverse()
    return w_vec

def correlation_from_covariance(cov_mat):
    v = np.sqrt(np.diag(cov_mat))
    outer_v = np.outer(v, v)
    corr_mat = cov_mat / outer_v
    corr_mat[cov_mat == 0] = 0
    return corr_mat

def pca_transform(df):
    w_vec = calc_weights(df)
    w_df = df.mul(w_vec, axis=0)
    cov_mat = np.dot(w_df.T, w_df)
    corr_mat = correlation_from_covariance(cov_mat)

    eig_vals, eig_vecs = np.linalg.eig(corr_mat)
    idxs = np.argsort(eig_vals)[::-1]
    eig_vals = eig_vals[idxs]
    eig_vecs = eig_vecs[idxs]

    top_eig_vals = eig_vals[:3]
    print(top_eig_vals)
    top_eig_vecs = eig_vecs[:3]

    cap_variation = top_eig_vals/np.sum(eig_vals)
    fig, ax = plt.subplots(figsize=(10,5))
    ax.bar([f'PC{i}' for i in range(top_eig_vals.shape[0])], cap_variation)
    plt.show()

    return top_eig_vals, top_eig_vecs

def create_eigenportfolios(top_eig_vecs, cols):
    pc1 = abs(top_eig_vecs[0])
    pc1 = pc1/pc1.sum()
    fig, ax = plt.subplots(figsize = (100,5))
    ax.bar(cols, pc1)
    plt.show()

    return pc1

```

```python
top_eig_vals, top_eig_vecs = pca_transform(train_df)
```

```python
weight_vec = create_eigenportfolios(top_eig_vecs, df.columns)
```

```python
def sharpe_ratio(ts_returns, periods_per_year=252):
    '''
    Sharpe ratio is the average return earned in excess of the risk-free rate per unit of volatility or total risk.
    It calculares the annualized return, annualized volatility, and annualized sharpe ratio.
    
    ts_returns are  returns of a signle eigen portfolio.
    '''
    n_years = ts_returns.shape[0]/periods_per_year
    annualized_return = np.power(np.prod(1+ts_returns),(1/n_years))-1
    annualized_vol = ts_returns.std() * np.sqrt(periods_per_year)
    annualized_sharpe = annualized_return / annualized_vol

    return annualized_return, annualized_vol, annualized_sharpe


def backtest(df, weights):
    portfolio_rets = np.dot(df, weights)
    ret, vol, sharpe = sharpe_ratio(portfolio_rets)
    print(ret, vol, sharpe)

    cum_returns = (1+portfolio_rets).cumprod()-1

    equal_weighted = np.ones(train_df.shape[1])/train_df.shape[1]
    equal_rets = np.dot(df, equal_weighted)
    cum_returns_equal = (1+equal_rets).cumprod()-1

    fig, ax = plt.subplots(figsize = (10, 5))
    ax.plot(df.index, cum_returns)
    ax.plot(df.index, cum_returns_equal)
    plt.legend()
    plt.show()
```
